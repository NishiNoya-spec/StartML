```python
df_1 = df.copy()
df_1['datetime'] = pd.to_datetime(df_1['datetime'])

print(
    f"""this is datetime data type: {df_1.loc[0, 'datetime']}: {type(df_1.loc[0, 'datetime'])}\n"""
    f"""this is string data type: {df.loc[0, 'datetime']:} {type(df.loc[0, 'datetime'])}\n"""
)
```

```python
df_meta = df_1.groupby(
    df_1['datetime'].dt.month
).agg({
    'temp': 'mean',
    'humidity': 'min',
    'registered': 'mean'
})

df_meta
```

```python
df_meta['label_registered'] = df_meta['registered'].apply(lambda x: 1 if x > 100.0 else 0)
df_meta
```

```python
df_2 = df.copy()
df_2['datetime'] = pd.to_datetime(df_2['datetime'])
df_temp = df_2.groupby(
    df_2['datetime'].dt.isocalendar().week
).agg({
    'temp': 'mean'
})
df_temp
```

```python
df_temp.plot(
    xlabel = 'Номер недели',
    ylabel = 'Среднее по температуре',
    title = 'Агрегирование по температуре', 
    grid = True, 
    figsize=(10, 5)
)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df_temp)
plt.title('Агрегирование по температуре', color='grey')
plt.xlabel('Номер недели')
plt.ylabel('Среднее по температуре')
plt.grid(True, alpha=0.4, color='black')

plt.show()
```
