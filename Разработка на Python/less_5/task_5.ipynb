```python
data = df.copy()
data['dateAdded'] = pd.to_datetime(data['dateAdded'])
data['dateUpdated'] = pd.to_datetime(data['dateUpdated'])
data['update_delta'] = (data['dateUpdated'] - data['dateAdded']).dt.days
mean_val = data['update_delta'].mean()
median_val = data['update_delta'].median()
print(mean_val, median_val)
```

```python
data['num_categories'] = data['categories'].str.split(',').apply(lambda x: len(set(x)) if isinstance(x, list) else 0)
filtered_data = data[data['num_categories'] > 20].copy()
filtered_data = filtered_data.groupby(
    [
        filtered_data['province']
    ], as_index=False
).agg({
    'longitude': 'min'
})
filtered_data['longitude'] = filtered_data['longitude'].apply(lambda x: round(x, 3))
filtered_data.to_csv('.\lesson5_task19.csv', sep=';', index=False)
```

```python
# Отберем записи с более чем 20 категориями
filtered_df = df[df['categories'].apply(lambda x: len(x.split(','))) > 20]

# Группируем по провинциям и находим минимальное значение longitude
result_df = filtered_df.groupby('province')['longitude'].min().reset_index()

# Округляем значения longitude до 3 знаков после запятой
result_df['longitude'] = result_df['longitude'].round(3)

# Сохраняем результат в CSV файл с сепаратором ;
result_df.to_csv('lesson5_task19.csv', sep=';', index=False)
```
