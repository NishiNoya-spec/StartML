```python
file_path = '.\\playstore.csv'
playstore = pd.read_csv(file_path, sep=',')
playstore
```

```python
unique_apps = playstore[playstore.duplicated(subset='App', keep=False)]
unique_apps.groupby('App').agg({'App': 'count'})
```

```python
playstore['App'].shape[0]
```

```python
playstore.drop_duplicates(subset='App', keep='first', inplace=True, ignore_index=True)
playstore
```

```python
column_list = list(playstore.columns)
ch_columns = [col.replace(' ', '_').lower() for col in column_list]
playstore.rename(columns=dict(zip(column_list, ch_columns)), inplace=True)
playstore.columns
```

```python
playstore
```

```python
print(
    f"""
        {type(playstore.loc[1, 'price'])}\n
        {type(playstore.loc[1, 'rating'])}\n
        {type(playstore.loc[1, 'reviews'])}\n
    """
)
```

```python
playstore['price'] = pd.to_numeric(playstore['price'].replace('[\$,]', '', regex=True), errors='coerce')

playstore_pivot_table = pd.pivot_table(

    playstore, 

    # группировка по признакам 
    index=[
        'category', 
        'type'
    ], 

    # признаки, по которым будет производится агрегирвоание данных 
    values=[
        'price',
        'rating',
        'reviews'
    ], 

    # функции агрегирования, которые будут применены
    aggfunc={
        'price': "mean",
        'rating': "mean", 
        'reviews': "mean"
    }

)

playstore_pivot_table = playstore_pivot_table.rename(columns={
    'price': 'mean_price', 
    'rating': 'mean_rating',
    'reviews': 'mean_reviews'
})

playstore_pivot_table['mean_price'] = round(playstore_pivot_table['mean_price'], 2)
playstore_pivot_table['mean_rating'] = round(playstore_pivot_table['mean_rating'], 1)
playstore_pivot_table['mean_reviews'] = round(playstore_pivot_table['mean_reviews'], 2)

playstore_pivot_table
```

```python
playstore_pivot_table.to_csv('.\\mini_proj_less5_task12.csv', sep=',', index=True)
result = pd.read_csv('.\\mini_proj_less5_task12.csv', sep=',')
result
```

```python
result.set_index(['category', 'type'], inplace=True)
result
```

```python
result.reset_index(['type'], inplace=True)
result
```

```python
playstore_table = playstore.groupby(
    [
        playstore['category'], 
        playstore['type']
    ]
).agg({
    'price': 'mean',
    'rating': 'mean',
    'reviews': 'mean' 
})

playstore_table
```

```python
playstore_table.reset_index(['type'], inplace=True)
playstore_table
```

```python
playstore_table_free = playstore_table[playstore_table['type'] == 'Free']
playstore_table_free = playstore_table_free.drop('type', axis=1)
playstore_table_paid = playstore_table[playstore_table['type'] == 'Paid']
playstore_table_paid = playstore_table_paid.drop('type', axis=1)
```

```python
playstore_table_paid_price = playstore_table_paid.drop(['rating', 'reviews'], axis=1)
playstore_table_paid_price
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

plt.figure(figsize=(10, 5))
plt.plot(playstore_table_paid_price)
plt.xlabel('Категория')
plt.ylabel('Цена')
x_values = playstore_table_paid_price.index
plt.xticks(rotation=45, ha='right')
plt.show()
```
