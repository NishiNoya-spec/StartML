```python
def func(x):
    return x**3 - 6*x**2 - 15*x + 10
```

```python
x = np.linspace(-5, 10, 150)
```

```python
fig = plt.figure()

fig.set_size_inches(10, 5)

plt.plot(x, func(x))

plt.xlabel('x')
plt.ylabel('f')
```

```python
def func_derivative(x):
    return 3 * x**2 - 12 * x - 15
```

```python
def descent_and_plot(X_start=10, learning_rate=0.01, threshold=0.0001):
    X_start = X_start
    learning_rate = learning_rate
    threshold = threshold

    fig = plt.figure()

    fig.set_size_inches(10, 5)

    plt.plot(x, func(x), alpha=0.5)

    plt.xlabel('x')
    plt.ylabel('f')

    a = []
    b = []

    while True:

        X_next = X_start - learning_rate * func_derivative(X_start)

        a.append([round(X_start, 2), round(X_next, 2)])
        b.append([round(func(X_start), 2), round(func(X_next), 2)])

        plt.scatter([X_start, X_next],
            [func(X_start), func(X_next)],
            color='green',
            s=80)

        plt.plot([X_start, X_next],
            [func(X_start), func(X_next)],
            color='purple')

        if abs(X_start - X_next) < threshold:
            break
        else:
            X_start = X_next


    print(f'       X={X_next}')
    print(f"   X: {a}")
    print(f"    X: {b}")

descent_and_plot()
```
